# Concatenate all Dataframes to One Table
- clean data to make values consistent for database load

```python
frames = [organized_1819_df, organized_1718_df, organized_1617_df, organized_1516_df, organized_1415_df, organized_1314_df, organized_1213_df, organized_1112_df, organized_1011_df]
joined_fedloan_df= pd.concat(frames, sort=False)
joined_fedloan_df = joined_fedloan_df.fillna(0)
joined_fedloan_df["ope_id"] = joined_fedloan_df["ope_id"].astype(int)
joined_fedloan_df["school"] = joined_fedloan_df["school"].str.title()
joined_fedloan_df["state"] = joined_fedloan_df["state"].astype(str)
joined_fedloan_df["zip_code"] = joined_fedloan_df["zip_code"].astype(str)
joined_fedloan_df["school_type"] = joined_fedloan_df["school_type"].str.title()
```

```python
joined_fedloan_df["school_type"] = joined_fedloan_df["school_type"].replace('Private', 'Private-Nonprofit')
joined_fedloan_df["school_type"] = joined_fedloan_df["school_type"].replace('Foreign Public', 'Foreign-Public')
joined_fedloan_df["school_type"] = joined_fedloan_df["school_type"].replace('Foreign Private', 'Foreign-Private')
joined_fedloan_df["school_type"] = joined_fedloan_df["school_type"].replace('Foreign For-Profit', 'Foreign-For-Profit')
```

```python
joined_fedloan_df = joined_fedloan_df.sort_values(by=["ope_id","year_begin"], ascending=[True,False])
joined_fedloan_df.head()
```

```python
uniqueschoolcodes = []
for schoolid in joined_fedloan_df["ope_id"]:
    if schoolid not in uniqueschoolcodes:
        uniqueschoolcodes.append(schoolid)

uniqueschoolnames = []
for code in uniqueschoolcodes:
    lookupresult = joined_fedloan_df.loc[joined_fedloan_df["ope_id"] == code, 'school'].iloc[0]
    uniqueschoolnames.append(lookupresult)
```

```python
schooltuples = list(zip(uniqueschoolcodes, uniqueschoolnames))
schoolname_df = pd.DataFrame(schooltuples, columns = ['ope_id', 'school_name_org'])
schoolname_df.head()
```

```python
yeargrouped = joined_fedloan_df.groupby(["aid_year"])
```

```python
yrdlsb_recip = yeargrouped["dlsb_recip"].sum()
yrugdlus_recip = yeargrouped["ugdlus_recip"].sum()
yrgrdlus_recip = yeargrouped["grdlus_recip"].sum()
yrprplus_recip = yeargrouped["prplus_recip"].sum()
yrgrplus_recip = yeargrouped["grplus_recip"].sum()
yrgrdlsb_recip = yeargrouped["grdlsb_recip"].sum()
yearlyrecip_df = pd.DataFrame({"total_dlsb_recip" : yrdlsb_recip,
                              "total_ugdlus_recip" : yrugdlus_recip,
                              "total_grdlus_recip" : yrgrdlus_recip,
                              "total_prplus_recip" : yrprplus_recip,
                              "total_grplus_recip" : yrgrplus_recip,
                              "total_grdlsb_recip" : yrgrdlsb_recip})
yearlyrecip_df = yearlyrecip_df.reset_index()
yearlyrecip_df
```

```python
yrdlsb_sum = yeargrouped["sum_dlsb_disb"].sum()
yrugdlus_sum = yeargrouped["sum_ugdlus_disb"].sum()
yrgrdlus_sum = yeargrouped["sum_grdlus_disb"].sum()
yrprplus_sum = yeargrouped["sum_prplus_disb"].sum()
yrgrplus_sum = yeargrouped["sum_grplus_disb"].sum()
yrgrdlsb_sum = yeargrouped["sum_grdlsb_disb"].sum()
yearlysum_df = pd.DataFrame({"total_dlsb_sum" : yrdlsb_sum,
                              "total_ugdlus_sum" : yrugdlus_sum,
                              "total_grdlus_sum" : yrgrdlus_sum,
                              "total_prplus_sum" : yrprplus_sum,
                              "total_grplus_sum" : yrgrplus_sum,
                              "total_grdlsb_sum" : yrgrdlsb_sum})
yearlysum_df = yearlysum_df.reset_index()
yearlysum_df
```

```python
publicschool_df = joined_fedloan_df.loc[joined_fedloan_df["school_type"] == "Public",:]
publiccount = publicschool_df["aid_year"].value_counts()

privateschool_df = joined_fedloan_df.loc[joined_fedloan_df["school_type"] == "Private-Nonprofit",:]
privatecount = privateschool_df["aid_year"].value_counts()

propschool_df = joined_fedloan_df.loc[joined_fedloan_df["school_type"] == "Proprietary",:]
propcount = propschool_df["aid_year"].value_counts()

forpubschool_df = joined_fedloan_df.loc[joined_fedloan_df["school_type"] == "Foreign-Public",:]
forpubcount = forpubschool_df["aid_year"].value_counts()

forprvschool_df = joined_fedloan_df.loc[joined_fedloan_df["school_type"] == "Foreign-Private",:]
forprvcount = forprvschool_df["aid_year"].value_counts()

forpropschool_df = joined_fedloan_df.loc[joined_fedloan_df["school_type"] == "Foreign-For-Profit",:]
forpropcount = forpropschool_df["aid_year"].value_counts()
```

```python
yearlyschoolct_df = pd.DataFrame({"school_count" : yeargrouped["school_type"].count(),
                                  "public" : publiccount,
                                 "private" : privatecount,
                                 "proprietary" : propcount,
                                 "foreign_public" : forpubcount,
                                 "foreign_private" : forprvcount,
                                 "foreign_profit" : forpropcount,
                                 "public_pct" : ((publiccount/yeargrouped["school_type"].count())*100).map("{:.2f}".format).astype(float),
                                 "private_pct" : ((privatecount/yeargrouped["school_type"].count())*100).map("{:.2f}".format).astype(float),
                                 "proprietary_pct" : ((propcount/yeargrouped["school_type"].count())*100).map("{:.2f}".format).astype(float),
                                 "forpub_pct" : ((forpubcount/yeargrouped["school_type"].count())*100).map("{:.2f}".format).astype(float),
                                 "forprv_pct" : ((forprvcount/yeargrouped["school_type"].count())*100).map("{:.2f}".format).astype(float),
                                 "forprop_pct" : ((forpropcount/yeargrouped["school_type"].count())*100).map("{:.2f}".format).astype(float)
                                 })
yearlyschoolct_df.index.name = "aid_year"
yearlyschoolct_df = yearlyschoolct_df.reset_index()
yearlyschoolct_df
```
